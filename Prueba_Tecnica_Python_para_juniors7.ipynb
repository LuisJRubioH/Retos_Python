# **Pruebas técnicas con python -- SySCursos (You Tube)**

```python
# En la variable 'text' dispones de una cadena de texto
# Debes encontarr las palabras y devolver cuántas veces se repite cada una de ellas
# Ejwemplo --- 'nadie' aperece 2 veces

text = "Creo que aveces es la gente de la que nadie espera nada, la quew hace cosas que nadie puede imaginar."
new_text = text.replace(",","").replace(".","").lower()

palabras = new_text.split()
palabra_unicas = []
count = 0
for palabra in palabras:
  if palabra not in palabra_unicas:
    palabra_unicas.append(palabra)
    count = palabras.count(palabra)
    print(f"'{palabra}' aparece {count} veces")

```

# ***Ejercicio  #2***

```python
#1. comprueba que los número en la lsita sean pares
#2. Encuentra los números pares únicos de la lista
#3. Suma estos números pares unicos de la lsiat y muestra por consola

numbers = [1,2,3,4,5,6,12,9,8,25,32,2,4,6]
```

**Mi solución**

```python
def identificar_pares(lista:list):
  for num in lista:
    if num % 2 == 0:
      print(True)
    else:
      print(False)

def encontra_pares_unicos(lista:list):
  pares_unuicos = [ ]
  for num in lista:
    if num % 2 == 0 and num not in pares_unuicos:
      pares_unuicos.append(num)
  return sorted(pares_unuicos)

def suma_de_pares_unicos(lista:list):
    return sum(encontra_pares_unicos(lista))

print(identificar_pares(numbers))
print(f"Los pares únicos son: {encontra_pares_unicos(numbers)}")
print(f" la suma de los pare súnicos es: {suma_de_pares_unicos(numbers)}")
```

# **Solución de SySCursos (You Tube)**

```python
def is_even_number(number):
  return number % 2 == 0

def find_unique_even_numbers(numbers):
  unique_even_numbers = set()
  for number in numbers:
    if is_even_number(number):
      unique_even_numbers.add(number)
  return unique_even_numbers

def sum_numbers(numbers):
  return sum(numbers)

find_unique_even_numbers = find_unique_even_numbers(numbers)
result = sum_numbers(find_unique_even_numbers)
print(f"La suma de los números pares y únicos de la lsita es: {result}")
```
